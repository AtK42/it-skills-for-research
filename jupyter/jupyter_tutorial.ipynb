We can easily display pandas dataframes and plot them:

````python
import numpy as np
import pandas as pd

# Generate a dataframe of uncorrelated standard normals
df = pd.DataFrame(np.random.standard_normal((1000, 3)), columns=["A", "B", "C"])

df.head()
````

````python
df.cumsum().plot(figsize=(10, 6), lw=2.25)
````

#### 5. Markdown

Markdown renders the code you type into a cell into HTML, allowing for a rather rich editing: bullet points, numbered lists, LaTeX equations and so on.

In a jupyter notebook while in command mode press `M` to convert a cell into markdown type and press `Y` to convert a cell into code type.

##### LaTeX equations

Inline expressions are typed in with single dollar sign delimiters: $e^{i\pi} = -1$ is an inline equation with raw command `$e^{i\pi} = -1$`

Double dollar sign delimiters are used for standalone equations:

$$\int_{a}^{b} f^{\prime} (x) dx = f(b) - f(a)$$


the source code of which is `$$\int_{a}^{b} f^{\prime} (x) dx = f(b) - f(a)$$`
